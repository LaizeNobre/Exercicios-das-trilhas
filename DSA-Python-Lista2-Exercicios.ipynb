#### Exercício 1

Escreva uma função que receba os valores de três lados de um triângulo e o classifique como "Equilátero" (todos os lados iguais), "Isósceles" (dois lados iguais) ou "Escaleno" (todos os lados diferentes).

```python
# Solução
lado1=float(input("digite a medida do primeiro lado:"))
lado2=float(input("digite a medida do segundo lado:"))
lado3=float(input("digite a medida do terceiro lado:"))

if lado1!=lado2 and lado1!=lado3 and lado2!=lado3:
    print("O triângulo é do tipo escaleno.")
elif lado1==lado2 and lado2==lado3 and lado3==lado1:
    print("O triângulo é do tipo equilátero.")
else:
    print("O triângulo é do tipo isóceles.")
```

#### Exercício 2

Escreva uma função que recebe um número inteiro e exibe a tabuada de multiplicação desse número, do 1 ao 10.

#### Exercício 3 

Você recebeu um dicionário com os nomes dos alunos e suas respectivas notas. Escreva uma função que calcula a média da turma e retorna uma lista com os nomes dos alunos que tiveram nota acima da média.

```python
# Solução
turma ={
    "pedro":7.5,
    "mariana":9.3,
    "eduarda":8.5,
    "gabriel":4.8,
    "vinicius":9.9,
    "camila":5.9
}
def calcular_media (turma):
    soma=0
    for aluno,nota in turma.items():
        soma+=nota
    media_turma=soma/len(turma.values())
    alunos={}
    for aluno,nota in turma.items():
        if nota> media_turma:
            alunos[aluno]=nota
    return f"Média da turma:{media_turma:.2f} \n alunos acima da média:{alunos}"

print(calcular_media(turma))
```

#### Exercício 5

Crie uma função chamada dsa_calcula_imc que aceite dois argumentos: peso (em kg) e altura (em metros). A função deve calcular o Índice de Massa Corporal (IMC) usando a fórmula IMC=peso/altura^2 e retornar o valor do IMC.

```python
# Solução
def dsa_calcula_imc(peso,altura):
    return peso/(altura*2)
peso=float(input("Digite seu peso em Kg:"))
altura=float(input("digite sua altura em metros:"))
print(f" Seu IMC é {dsa_calcula_imc(peso,altura):.2f}")
```

#### Exercício 6

Você tem uma lista de dicionários, onde cada dicionário representa uma pessoa com nome e idade. Use a função sorted() com uma expressão lambda como chave (key) para ordenar a lista de pessoas da mais nova para a mais velha.

```python
# Solução
pessoas = [
    {"nome": "Ana", "idade": 30},
    {"nome": "Bruno", "idade": 25},
    {"nome": "Carlos", "idade": 40},
    {"nome": "Diana", "idade": 20}
]
lista_ordenada=[sorted(pessoas, key=lambda pessoa: pessoa["idade"])]
print(lista_ordenada)
```

#### Exercício 7

Crie uma função que receba uma lista de números inteiros e retorne um dicionário contendo a contagem de quantos números são pares e quantos são ímpares.

```python
# Solução
numeros=[]
while True:
    n=int(input("digite um número inteiro ou negativo para parar:"))
    if n<0:
        print("incrementação encerrada")
        break
    numeros.append(n)
def verificar_lista(lista):
    resultado={}
    resultado["pares"]=[x for x in lista if x%2==0]
    resultado["impares"]=[x for x in lista if x%2!=0]
    return resultado.items()

print(verificar_lista(numeros))
```

#### Exercício 8

Crie uma função que receba uma lista de strings (potenciais e-mails) e um parâmetro opcional dominio_desejado (com valor padrão "gmail.com"). A função deve retornar uma nova lista, usando list comprehension, contendo apenas os e-mails que terminam com o domínio desejado.

```python
# Solução
def filtrar_emails(lista_emails,dominio_desejado="gmail.com"):
    return [ email for email in lista_emails if email.endswith("@"+dominio_desejado)]

lista=[]
while True:
    l=input("digite um potencial email ou 0 para sair:")
    if l=="0":
        break
    lista.append(l)
print(filtrar_emails(lista))
```

#### Exercício 9

Dada uma lista de frases, use a função map() em conjunto com uma expressão lambda para criar uma nova lista onde cada frase é convertida para letras maiúsculas e tem a palavra "PYTHON" anexada ao final.

```python
# Solução
lista=["py é legal","programar é interessante","estudos na dsa"]
lista_nova=list(map(lambda frase: frase.upper() +" PYTHON",lista))
print(lista_nova)
```

#### Exercício 10

Crie um jogo onde o computador escolhe um número secreto entre 1 e 20. O jogador tem 5 tentativas para adivinhar. A cada tentativa, o programa informa se o palpite foi muito alto ou muito baixo. Se o jogador acertar, o loop deve ser interrompido com uma mensagem de vitória.

```python
# Solução
import random
numero_secreto=random.randint(1,20)
tentativas=5
print("Pense em um número de 1 a 20")
print("Você tem 5 chances de tentar acertar o número secreto:")
for tentativa in range(1,tentativas+1):
    palpite=int(input(f"tentativa {tentativa}: digite seu palpite:"))
    if palpite ==numero_secreto:
        print(f"Parabéns! O número secreto é {palpite}")
        print("Fim de jogo!")
        break
    elif palpite>numero_secreto:
        print(f"O número {palpite} é maior que o numero secreto")
    elif palpite<numero_secreto:
        print(f"O número {palpite} é menor que o numero secreto")
    else:
        print(f"Ops, tente novamente!")
else:
    print("Suas tentativas acabaram. Boa sorte na próxima!")
```
